```python
#Cleaning - Rename Columns
df.rename(columns={'New Georeferenced Column' : 'Georeferenced'})
```

```python
#Rename Boroughs
df['ARREST_BORO'].replace('M','Manhattan',inplace=True)
df['ARREST_BORO'].replace('Q','Queens',inplace=True)
df['ARREST_BORO'].replace('K','Brooklyn',inplace=True)
df['ARREST_BORO'].replace('S','Staten Island',inplace=True)
df['ARREST_BORO'].replace('B','Bronx',inplace=True)
```

```python
#Confirm Rename Successful
df.head()
```

```python
#Research Quesion: What is the highest felony crime committed and what boroughs attribute to that volume. 
```

```python
df1 = df[['ARREST_DATE', 'OFNS_DESC', 'LAW_CAT_CD', 'ARREST_BORO', 'AGE_GROUP', 'PERP_SEX', 'PERP_RACE']]
```

```python
sns.countplot(x = 'ARREST_BORO', hue = 'LAW_CAT_CD', data = df1)
```

```python
df2 = df1.loc[(df1.LAW_CAT_CD == 'F')]
```

```python
df2.groupby(['ARREST_BORO']).count()
```

```python
sns.countplot(x = 'ARREST_BORO', hue = 'LAW_CAT_CD', data = df2)
```

```python
df3 = df1.loc[(df1.LAW_CAT_CD == 'M')]
```

```python
df3.groupby(['ARREST_BORO']).count()
```

```python
sns.countplot(x = 'ARREST_BORO', hue = 'LAW_CAT_CD', data = df3)
```

```python
# The highest felony crime is 'ASSULT 3 & RELATED OFFENSES'
df2.groupby(['OFNS_DESC']).count()
```

```python
df4 = df2.loc[(df2.OFNS_DESC == 'ASSAULT 3 & RELATED OFFENSES')]
```

```python
df4.groupby(['ARREST_BORO']).count()
```

```python
sns.countplot(x = 'ARREST_BORO', hue = 'OFNS_DESC', data = df4)
```

```python
# In Summary the highest felony crime committed is 'ASSAULT 3 & RELATED OFFENSES' totaling 7888. The Bronx Borough has
# the most felonies committed with this offense. Brookly is a close second, with Staten Island having the least. 
```
